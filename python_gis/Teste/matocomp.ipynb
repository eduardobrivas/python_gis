```python
import geopandas as gpd
import pandas as pd
import fiona 

# Especificar o nome do conjunto de dados dentro do arquivo .gdb
nome_do_conjunto = 'UsoActual_BR'

# Caminho para o arquivo .gdb
gdb_file = 'D:\\ARAUCO\\Movimentos_Novembro2021.gdb'

# Construir o caminho completo para o conjunto de dados
caminho_completo = f'{gdb_file}:{nome_do_conjunto}'
fiona.listlayers(gdb_file)

# Carregar o conjunto de dados para um GeoDataFrame
gdf = gpd.read_file(gdb_file, layer= (nome_do_conjunto))

# Restante do código permanece o mesmo...

gdf.head()
```

```python
# Criar a chave combinando os campos 'predio', 'talhao' e 'mes'
gdf['CHAVE_'] = gdf['PREDIO'].astype(str) + '_' + gdf['TALHAO'].astype(str) + '_' + gdf['MESPLANTACION'].astype(str)


print(gdf)
gdf.crs
```

```python
# Carregar o arquivo do Excel para um DataFrame
excel_file = 'seuarquivo.xlsx'
df_excel = pd.read_excel(excel_file)

# Fazer a união entre o GeoDataFrame e o DataFrame usando a coluna chave
result = gdf.merge(df_excel, how='left', left_on='chave', right_on='sua_coluna_chave_do_excel')
```
